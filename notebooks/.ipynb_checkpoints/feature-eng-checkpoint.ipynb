```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from plotnine import *
```

```python
train_set = pd.read_csv('../data/application_train.csv')
test_set = pd.read_csv('../data/application_test.csv')
```

```python
def encode_and_align_dfs(dfs):
    le = LabelEncoder()
    le_count = 0
    trn_df = dfs[0]
    tst_df = dfs[1]
    for col in trn_df:
        if trn_df[col].dtype == 'object':
            # If 2 or fewer unique categories
            if len(list(trn_df[col].unique())) <= 2:
                # Train on the training data
                le.fit(trn_df[col])
                # Transform both training and testing data
                trn_df[col] = le.transform(trn_df[col])
                tst_df[col] = le.transform(tst_df[col])

                # Keep track of how many columns were label encoded
                le_count += 1
    print('%d columns were label encoded.' % le_count)
    print('\nBegining process to one-hot encode the rest...')
    trn_df = pd.get_dummies(trn_df)
    tst_df = pd.get_dummies(tst_df)
    print('Training Features shape: ', trn_df.shape)
    print('Testing Features shape: ', tst_df.shape)
    print('\nBegining process to align dataframes...')
    
    train_labels = trn_df['TARGET']

    # Align the training and testing data, keep only columns present in both dataframes
    trn_df, tst_df = trn_df.align(tst_df, join = 'inner', axis = 1)

    print('Training Features shape: ', trn_df.shape)
    print('Testing Features shape: ', tst_df.shape)
    trn_df['TARGET'] = train_labels
    
    return (trn_df, tst_df)
```

```python
train_set, test_set = encode_and_align_dfs((train_set, test_set))
```

```python
# Find correlations with the target and sort
correlations = train_set.corr()['TARGET'].sort_values()

# Display correlations
print('Most Positive Correlations: \n', correlations.tail(15))
print('\nMost Negative Correlations: \n', correlations.head(15))
```

```python
# Find the correlation of the positive days since birth and target
train_set['DAYS_BIRTH'] = abs(train_set['DAYS_BIRTH'])
train_set['DAYS_BIRTH'].corr(train_set['TARGET'])
```
